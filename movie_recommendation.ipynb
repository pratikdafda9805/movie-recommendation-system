```python
import ast
```

```python
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l    
```

```python
def convert3(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if(counter!=3):
            l.append(i['name'])
            counter+=1
        else:
            return l
             
```

```python
movies['keywords'] = movies['genres'].apply(convert)
```

```python
def convert_crew(obj):
    l = []
    for i in ast.literal_eval(obj):
        if(i['job']=='Director'):
            l.append(i['name'])
    return l        
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps =PorterStemmer()
```

```python
def stem(text):
    list=[]
    for i in text.split():
       list.append(ps.stem(i))
    return " ".join(list)
```

```python
movies['tags']=movies['tags'].apply(stem)
```

```python
cv.fit_transform(movies['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(obj):
    movie_index = movies[movies['title']==obj].index[0]
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(i[0])
        print(movies.iloc[i[0]].title)
```

```python
recommend("The Avengers")
```
